```python
import gravis as gv
```

```python
# Using gJGF, gravis JSON graph format.
graph1 = {
    'graph': {
        'directed': False,
        'metadata': {
            'background_color': 'black',
            'edge_size': 3,
            'edge_label_size': 14,
            'edge_label_color': 'white',
            'edge_color': '#66bd63',
            'node_size': 15,
            'node_color': 'white',
            'node_label_color': 'white',
        },
        'nodes': {
            1: {'metadata': {'title': 'John Smith', 'size': 5, 'color': 'red'}},
            2: {'metadata': {'title': 'Jane Smith', 'size': 2, 'click': '<B>alert</B>'}},
            3: {'metadata': {'title': 'Michael Smith', 'size': 2}},
            4: {'metadata': {'title': 'Kelly Smith', 'size': 2}},
            5: {'metadata': {'title': 'Karen Barnes', 'size': 1}},
            6: {'metadata': {'title': 'Richard Barnes', 'size': 1}},

        },
        'edges': [
            {'source': 1, 'target': 2, 'metadata': {'relationship': 'wife'}},
            {'source': 1, 'target': 3, 'metadata': {'relationship': 'son'}},
            {'source': 1, 'target': 4, 'metadata': {'relationship': 'daughter'}},
            {'source': 2, 'target': 5, 'metadata': {'relationship': 'mother'}},
            {'source': 2, 'target': 6, 'metadata': {'relationship': 'father'}},
        ],
    }
}
```

```python
gv.vis(graph1, 
       show_node_label=True, 
       node_label_data_source='title', 
       node_drag_fix=True,
       #node_size_data_source='size',
       node_size_factor=10,
       show_edge_label=True, 
       edge_label_data_source='relationship',
      edge_curvature=0.3)
```

```python
gv.d3(graph1,
        node_label_data_source='title', 
       node_drag_fix=True,
       node_size_data_source='size',
       node_size_factor=10,
       show_edge_label=True, 
       edge_label_data_source='relationship',     
     edge_curvature=0.3)
```
